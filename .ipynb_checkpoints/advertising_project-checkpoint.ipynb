# The machine learning problem

We are statistical consultants hired by a client to provide advice on how to improve sales of a particular product. The Advertising data set consists of the sales of that product in 200 different markets, along with advertising budgets for the product in each of those markets for three different media: TV, radio, and newspaper.

1) Our goal is to develop an accurate model that can be used
to predict sales on the basis of the three media budgets.

## We begin by loading/imprting our data into python.

```python

import pandas as pd

filename = 'http://faculty.marshall.usc.edu/gareth-james/ISL/Advertising.csv'
advert = pd.read_csv(filename,index_col=0)
index = advert.index
columns = advert.columns
values = advert.values
print(advert.shape)

```

Let's split pur data into training and validation sets:

```python
from sklearn.model_selection import train_test_split
import numpy as np
X = advert[['TV','radio','newspaper']]#use double square brackets for multiple columns
y= advert['sales'] #use single bracket for a single column
#selecting a column resturs a series and not a dataframe
#a way around this is to do advert[['sales']], this returns a dataframe
X.shape #x is a pandas dataframe while y is just a series.

Y=y[:,np.newaxis]# turn y into (n_samples,_n_features) and not a series
#Y.shape
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y,random_state=1, test_size = 0.5) #split data into training (50%) and test (50%)
```

Next, we pick the algorithms we want to apply and fit them to the data.
After that, we see how they perform on test data.


```python
from sklearn.model_selection import cross_validate,KFold,StratifiedKFold
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
kfold = KFold(n_splits=10) #maintains proportions in the data
scores = cross_validate(reg, Xtrain, ytrain, cv=kfold,scoring = 'neg_mean_squared_error',return_estimator = True)
```

Let's see how our model looks like and what coefficients it has

```python
for model in scores['estimator']:
    
    print(model.coef_)
    
for model in scores['test_score']:
    print(-1*model) 
    
scores['test_score'].mean()
#It appears that the second last model is the best one, low error.
```

Fit the model

```python
reg.fit(Xtrain,ytrain)
print("model.coef_: {}".format(reg.coef_))
print("model.intercept_: {}".format(reg.intercept_))
```

Let's see how it predicts the test set.


```python
y_hat = reg.predict(Xtest)
from sklearn.metrics import mean_squared_error #ch
mean_squared_error(ytest, y_hat)
```

## Using trees to predict sales

1) A tree without prepruning

```python
from sklearn.tree import DecisionTreeRegressor
salest = DecisionTreeRegressor(random_state=0)

tree_scores = cross_validate(salest,Xtrain,ytrain,cv=kfold,scoring = 'neg_mean_squared_error',return_estimator = True)

```

```python
for model in tree_scores['test_score']:
    print(-1*model) 
    
tree_scores['test_score'].mean()*-1
```

Let's see how well it does on test data:

```python
salest = salest.fit(Xtrain,ytrain)
predicteds = salest.predict(Xtest)

mean_squared_error(ytest, predicteds)
```

```python
#A good way to visualize tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(25,10))
a = plot_tree(salest, 
              feature_names= ['TV','radio','newspaper'],  
              filled=True, 
              rounded=True, 
              fontsize=14)
```

```python
def plot_feature_importances(model):
    n_features = 3
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), ['TV','radio','newspaper'])
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    
    
plot_feature_importances(salest)
```

It appears that the TV budget seems to be response for almost all the predictive power.

Let's preprune tree:

```python
salest = DecisionTreeRegressor(max_depth= 5,max_leaf_nodes = 10, random_state = 1)
salest = salest.fit(Xtrain,ytrain)
predicteds = salest.predict(Xtest)

mean_squared_error(ytest, predicteds)
```

Messing with the parameters seems to make tree worse.

2) Random forest graph

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(random_state = 2, max_features = 'log2')
rf = rf.fit(Xtrain,ytrain)
rfpredicts = rf.predict(Xtest)
mean_squared_error(ytest, rfpredicts)
```

```python
plot_feature_importances(rf)
```

```python
Unprunned tree does sligtly better than random forsest.
```

3) Boosted tree

```python
from sklearn.ensemble import GradientBoostingRegressor
booster = GradientBoostingRegressor(random_state=3 ,max_depth = 3,learning_rate = 0.1)
booster = booster.fit(Xtrain,ytrain)
boostpredicts = booster.predict(Xtest)
mean_squared_error(ytest, boostpredicts)
```

```python
plot_feature_importances(booster)
```

# Conclusion

It is clear that the boosted model does better than all the models considered in terms of predictiong sales. So, it should be used. Also, sales seem to be dertemined primarily by the TV budget with the radio budget following it. The proportions are around 60% TV and 38% radio.
